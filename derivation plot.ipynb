```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# input
def function(x):
    f = np.sqrt(np.sinh(x))/np.tan(x**3) + (x**10)*np.sin(x)
    return f
start = 1
end = 5
iteration = 0.01
mode = 10
```

```python
x = [i for i in np.arange(start, end, iteration)]
func = [function(i) for i in x]
def derivative(i, list):
    f_iplush = list[i+1]
    f_iminush = list[i-1]
    return (f_iplush - f_iminush)/(2*iteration)
def derivation_list(limit_range, list):
    li = []
    for i in limit_range:
        li.append(derivative(i, list))
    return li
d_list = [
    [x, func],
]
mode_reverse = 1
for i in range(mode):
    if mode_reverse > mode:
        break
    limit_range = range(1, len(d_list[mode_reverse-1][1])-1)
    x_range = x[mode_reverse:len(x)-mode_reverse]
    d_list.append([x_range, derivation_list(limit_range, d_list[mode_reverse-1][1])])
    mode_reverse += 1
```

```python
i = 1
plt.figure(figsize=(22, mode*3))
for li in d_list:
    plt.subplot(int(len(d_list)/3)+2, 3, i)
    plt.plot(li[0], li[1], color='red')
    plt.title(f'derivation mode {i-1}')
    plt.grid()
    i += 1
plt.show()
```

```python
i=1
plt.figure(figsize=(22, 11))
for li in d_list:
    plt.plot(li[0], li[1])
    plt.title('all in one')
    plt.grid()
    i += 1
plt.show()
```
